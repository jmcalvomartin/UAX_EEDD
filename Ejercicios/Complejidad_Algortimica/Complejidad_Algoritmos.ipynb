# Complejidad Algorítmica

### Definición
La **complejidad algorítmica** se refiere a la medida de la cantidad de recursos computacionales necesarios para ejecutar un algoritmo. Estos recursos pueden ser tiempo de ejecución o espacio de memoria requerido. La complejidad algorítmica se utiliza para evaluar el rendimiento de los algoritmos y determinar su eficiencia en términos de consumo de recursos.

La complejidad algorítmica se expresa generalmente en función del tamaño de entrada del algoritmo, denotado como **"n"**. La notación **Big O** se utiliza para describir la complejidad algorítmica de manera asintótica, lo que significa que se enfoca en el comportamiento del algoritmo a medida que el tamaño de entrada tiende hacia infinito. 

#### Ejemplo
Un algoritmo con complejidad **O(n)** indica que su tiempo de ejecución aumenta linealmente con el tamaño de entrada. Por otro lado, un algoritmo con complejidad $ O(n^2) $  indica que su tiempo de ejecución aumenta cuadráticamente con el tamaño de entrada. La complejidad algorítmica proporciona una herramienta importante para el análisis y diseño de algoritmos eficientes, permitiendo a los desarrolladores tomar decisiones informadas sobre qué enfoque utilizar para resolver un problema.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Función para representar una gráfica de complejidad
def plot_complexity(x, y, complexity_type):
    plt.plot(x, y, label=complexity_type)

# Crear datos de ejemplo
#x = np.linspace(1, 7, 100)  # Valores decimales de 1 a 10, con 100 puntos
x = np.arange(1, 8)
y_linear = x
y_quadratic = x**2
y_exponential = 2**x
y_logarithmic = x * np.log(x)
y_factorial = [np.math.factorial(int(n)) for n in x]

# Crear la gráfica
plt.figure()

# Graficar las complejidades en la misma figura
plot_complexity(x, y_linear, 'O(n)')
plot_complexity(x, y_quadratic, 'O(n^2)')
plot_complexity(x, y_exponential, 'O(2^n)')
plot_complexity(x, y_logarithmic, 'O(n * log(n))')
plot_complexity(x, y_factorial, 'O(n!)')

# Personalizar la gráfica
plt.xlabel('Tamaño de la entrada')
plt.ylabel('Tiempo de ejecución')
plt.legend()

# Mostrar la gráfica
plt.show()
```

## Ejemplos tipos de complejidad

### O(n): Lineal
* Un programa que me diga el número más alto de una lista

```python
import random

lista=[random.randint(0,100) for x in range(0,25)]
print(lista)
max=lista[0]
for x in range (0,len(lista)):
  if lista[x]>max:
    max=lista[x]
print("El número más alto es ",max)
```

### $O(n^2)$: Cuadrático
* Un programa que ordena una lista dada

```python
import random

lista=[random.randint(0,1000) for x in range(0,20)]
lista2=[]
print(lista)

while len(lista)>0:
    max=lista[0]
    for x in range (0,len(lista)):
      if lista[x]>max:
        max=lista[x]
    lista2.append(max)
    lista.remove(max)
print(lista2)
```

### $ O(n!) $ : El problema del viajante
El problema del viajante (también conocido como el problema del viajante de comercio) es un clásico problema de optimización que busca encontrar el recorrido más corto que pase por todos los nodos de un grafo, visitando cada nodo exactamente una vez y regresando al nodo de partida. La solución óptima implica recorrer todas las posibles combinaciones de nodos y calcular la distancia total para cada una.
<br>
La complejidad de este algoritmo es de **O(n!)**, donde n es el número de nodos en el grafo. Esto se debe a que se generan todas las permutaciones posibles, lo cual implica una explosión combinatoria a medida que aumenta el número de nodos. Para problemas con un número grande de nodos, este enfoque de fuerza bruta no es eficiente y se requieren algoritmos más sofisticados, como el algoritmo del vecino más cercano o el algoritmo del árbol de expansión mínima. Estos algoritmos ofrecen soluciones aproximadas en tiempos de ejecución más eficientes.

```python
import random

def create_random_graph(num_nodes):
    graph = {}

    # Generar pesos aleatorios para las aristas
    for i in range(num_nodes):
        graph[i] = {}

    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):
            weight = random.randint(1, 100)  # Asignar un peso aleatorio
            graph[i][j] = weight
            graph[j][i] = weight

    return graph
```

```python
import itertools
import networkx as nx
import matplotlib.pyplot as plt

def tsp(graph):
    nodes = list(graph.keys())
    permutations = list(itertools.permutations(nodes))

    min_distance = float('inf')
    optimal_path = None

    for path in permutations:
        distance = 0
        for i in range(len(path) - 1):
            current_node = path[i]
            next_node = path[i + 1]
            distance += graph[current_node][next_node]
        
        if distance < min_distance:
            min_distance = distance
            optimal_path = path
    
    return optimal_path, min_distance

# Ejemplo de grafo
graph = create_random_graph(6)

optimal_path, min_distance = tsp(graph)
print("Optimal Path:", optimal_path)
print("Minimum Distance:", min_distance)

# Crear y visualizar el grafo
G = nx.Graph(graph)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue')
labels = nx.get_edge_attributes(G,'color')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Grafo del problema del viajante')
plt.show()

# Dibujar la ruta óptima
optimal_edges = [(optimal_path[i], optimal_path[i+1]) for i in range(len(optimal_path)-1)]
optimal_edges.append((optimal_path[-1], optimal_path[0]))  # Añadir el último tramo
optimal_labels = {(optimal_path[i], optimal_path[i+1]): graph[optimal_path[i]][optimal_path[i+1]] for i in range(len(optimal_path)-1)}
optimal_labels[(optimal_path[-1], optimal_path[0])] = graph[optimal_path[-1]][optimal_path[0]]  # Añadir etiqueta del último tramo

plt.figure()
nx.draw(G, pos, with_labels=True, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edgelist=optimal_edges, edge_color='red', width=2.0)
nx.draw_networkx_edge_labels(G, pos, edge_labels=optimal_labels, font_color='red')
plt.title('Ruta óptima del problema del viajante')
plt.show()
```

## Mejorar algortimo: $ O(n) $ Vs $ O(long(n)) $

### Indicar la posición de un elemento en una lista ordenada
El objetivo de este ejercicio es mejorar el tiempo de computación, es decir la complejidad **(O(n))** de un algortimo para la búsqueda de elementos en una lista ordenada.


#### Busqueda Binaria
La complejidad de este algoritmo de búsqueda binaria es O(log n) debido a que en cada iteración, el tamaño del rango de búsqueda se reduce a la mitad. Sin embargo, si se realiza la búsqueda en una lista que ya está ordenada, la complejidad final sería O(n log n) porque se deben realizar n comparaciones (número máximo de iteraciones) en total para encontrar el elemento objetivo en la lista ordenada.

```python
import random
import time

#Utilizamos random.sample para generar una lista de números únicos dentro de un rango
my_list=random.sample(range(0,1000000),800000)
#Utilizamos la función sort para ordenar la lista 
my_list.sort()

print("Mi lista ordenada es: ")
#print(my_list)
search = int(input("Busqueda: "))


#Algortimo tradicional

iniciar=time.time()

for x in range (0,len(my_list)):
  if my_list[x]==search:
    print("Su posición es ", x)
    break

final=time.time()

print("El tiempo del algortimo tradicional ", final-iniciar)


#Algoritmo Binario
primero=0
ultimo=len(my_list)
mitad=0

iniciar=time.time()
while primero < ultimo:
  mitad = (primero + ultimo) // 2
  if my_list[mitad] > search:
    ultimo = mitad -1
  elif my_list[mitad] < search:
    primero = mitad + 1
  else:
    print("La posición es ", mitad)
    break
final=time.time()

print("El tiempo del algortimo binario ", final-iniciar)
```

# Operaciones con Matrices

### Manejo de las matrices por medio de Python

La aplicación directa de la definición matemática de multiplicación de matrices proporciona un algoritmo que requiere $n^3$ operaciones de campo para multiplicar dos **n × n** matrices sobre ese campo $O(n^3)$ (en notación O grande). <br>
Sorprendentemente, existen algoritmos que proporcionan mejores tiempos de ejecución que este sencillo **"algoritmo de libros de texto"**. 

<br>

El primero en ser descubierto fue **el algoritmo de Strassen** , ideado por Volker Strassen en 1969 y al que a menudo se hace referencia como **"multiplicación rápida de matrices"**. Aún se desconoce el número óptimo de operaciones de campo necesarias para multiplicar dos matrices cuadradas n × n hasta factores constantes . Ésta es una importante cuestión abierta en la informática teórica .

### Objetivo
* Crear un código que permita el producto de matrices (cuadradas o no cuadradas) y de como resultado una matriz y valorar su velocidad de ejecución según podamos ir haciendo matrices cada vez más grandes.

```python
# Multiplicar dos matrices de forma tradicional
def multiplicar_matrices(A, B):
  # Comprobamos que las matrices se puedan multiplicar
  if len(A[0]) != len(B):
    print("Las matrices no se pueden multiplicar")
    return

  # Creamos una matriz resultado con tantas filas como A y tantas columnas como B
  C = np.zeros([A.shape[0],B.shape[1]])

  # Recorremos las filas de A
  for i in range(len(A)):
    # Recorremos las columnas de B
    for j in range(len(B[0])):
      # Recorremos las filas de B
      for k in range(len(B)):
        # Vamos sumando los productos de cada elemento de la fila de A por cada elemento de la columna de B
        C[i][j] += A[i][k] * B[k][j]

  return C
```

## Algoritmo de Strassen

El algoritmo de Strassen es un algoritmo eficiente de multiplicación de matrices que le permite multiplicar dos matrices en un tiempo $ O(n^{2.807}) $, en lugar del tiempo $ O(n^3) $ que lleva multiplicar dos matrices utilizando el algoritmo estándar de multiplicación de matrices.

Esta implementación utiliza llamadas a funciones recursivas para dividir las matrices en submatrices más pequeñas hasta que alcanzan un tamaño de 1x1, momento en el que se pueden multiplicar directamente. Luego, los valores intermedios se calculan mediante una serie de sumas y restas de matrices, y el resultado final se ensambla combinando los valores intermedios.

<br>

Artículo completo en https://www.europeanvalley.es/noticias/multiplicacion-de-matrices-la-importancia-de-un-algoritmo/

```python
def strassen(A, B):
  n = len(A)

  if n == 1:
    return A * B
  else:
    # Divide the matrices into four submatrices
    a11 = A[:n//2, :n//2]
    a12 = A[:n//2, n//2:]
    a21 = A[n//2:, :n//2]
    a22 = A[n//2:, n//2:]
    b11 = B[:n//2, :n//2]
    b12 = B[:n//2, n//2:]
    b21 = B[n//2:, :n//2]
    b22 = B[n//2:, n//2:]

    #print (a11,a22)
    #print(a11+a22)

    # Compute intermediate values
    m1 = strassen(a11 + a22, b11 + b22)
    m2 = strassen(a21 + a22, b11)
    m3 = strassen(a11, b12 - b22)
    m4 = strassen(a22, b21 - b11)
    m5 = strassen(a11 + a12, b22)
    m6 = strassen(a21 - a11, b11 + b12)
    m7 = strassen(a12 - a22, b21 + b22)

    # Compute the final values of C
    c11 = m1 + m4 - m5 + m7
    c12 = m3 + m5
    c21 = m2 + m4
    c22 = m1 - m2 + m3 + m6

    # Assemble the final matrix
    C = np.zeros((n, n))
    C[:n//2, :n//2] = c11
    C[:n//2, n//2:] = c12
    C[n//2:, :n//2] = c21
    C[n//2:, n//2:] = c22


    return C
```

## Análisis de ambos algoritmos

```python
import numpy as np
import time
import pandas as pd
import plotly.express as px

dim=[4,5,6,7,8]
t_tradicional=[]
t_strassen=[]

#Bucle de pruebas

for n in dim:

    A = np.random.randint(1,100, size=(2**n,2**n)) #Se crear una matriz con valores aleatorios
    B = np.random.randint(1,100, size=(2**n,2**n)) #Se crear una matriz con valores aleatorios

    print("---------------Matrices cuadradas de dimensión ", A.shape, "-------------------------------------")
    #print(A)
    #print("\n")
    #print(B)
    #print("---------------------------")

    inicio=time.time()
    resultado=strassen(A,B)
    final=time.time()
    #print(resultado)
    t_strassen.append(final-inicio)
    print("Tiempo de realización algoritmo Strassen ", final-inicio)

    print("---------------------------------------------------------")

    inicio=time.time()
    resultado=multiplicar_matrices(A,B)
    final=time.time()
    #print(resultado)
    t_tradicional.append(final-inicio)
    print("Tiempo de realización algoritmo Tradicional ", final-inicio)

    print("\n")
    print("---------------------------------------------------------")

##Grafica de resultados
dimension=[]
for x in dim:
  dimension.append(2**x)

data={"Tiempo_Tradicional": t_tradicional,"Tiempo_Strassen": t_strassen, "Dimensiones":dimension}
      
df=pd.DataFrame(data, index=dimension)

fig = px.line(df[["Tiempo_Tradicional","Tiempo_Strassen"]], 
              labels={
                     "index": "Dimensiones",
                     "value": "Tiempos de ejecuación",
                    "variable" : "Tipo de Algoritmo"
                 }, title="Comparativa algoritmos Multiplicación Matrices", template="plotly_dark", text=dimension, markers=True)
fig.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_weighted_graph(graph):
    G = nx.Graph()

    # Agregar nodos al grafo
    G.add_nodes_from(graph.keys())

    # Agregar aristas con pesos
    for node, neighbors in graph.items():
        for neighbor, weight in neighbors.items():
            G.add_edge(node, neighbor, weight=weight)

    # Obtener posiciones para dibujar el grafo
    pos = nx.spring_layout(G)

    # Dibujar el grafo
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, edge_color='gray')
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Mostrar el grafo
    plt.title("Grafo con Pesos")
    plt.axis('off')
    plt.show()
```
